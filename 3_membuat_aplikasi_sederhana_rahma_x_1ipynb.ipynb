#aplikasi menghitung luas persegi panjang

```python
panjang = int(input("masukan panjang dari persegi panjang (cm):"))
lebar = int(input("masukan lebar dari persegi panjang (cm):"))

luas = panjang*lebar

print(f"\nluas dari persegi panjangnya adalah {luas} cm^2.")
```

#aplikasi menghitung volume balok

```python
panjang = int(input("Masukkan panjang dari balok (cm): "))
lebar = int(input("Masukkan lebar dari balok (cm): "))
tinggi = int(input("Masukkan tinggi dari balok (cm): "))

volume = panjang * lebar * tinggi

print(f"\nVolume dari baloknya adalah {volume} cm^3.")
```
